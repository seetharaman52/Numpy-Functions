```python
x = np.array([1,2,3,4])
y = np.array([5,6,7,8])
print("dot:",np.dot(x,y))
print("vdot:",np.dot(x,y))
```

```python
#In 2D
a = np.array([[1,2],[4,3]])    #1 2 | 6 7
b = np.array([[6,7],[5,8]])    #4 3 | 5 8
print("dot:\n",np.dot(a,b))
print("vdot: ",np.vdot(a,b))   #1x6+ 2x7+ 4x5 + 3x8
```

```python
#Inner and Outer element
a = np.array([1,2,3])
b = np.array([4,5,6])
print("Inner:",np.inner(a,b))
print("Outer:\n",np.outer(a,b))
#Inner --> 1x4 + 2x5 + 3x6
#Outer --> 1xb, 2xb, 3xb
```

```python
a = np.array([[1,2],[4,3]])     #1 2 | 6 7
b = np.array([[6,7],[5,8]])     #4 3 | 5 8
print("Inner:\n",np.inner(a,b))
print("Outer:\n",np.outer(a,b)) #Inner --> 1x6 + 2x7 = 20; 1x5 + 2x8 = 21; 4x6 + 3x7 = 45; 4x5 + 3x8 = 44  
```

```python
#Determinant
a = np.array([[1,2],[-3,4]])
print("Determinant:", np.linalg.det(a))
# --> 1x4 - (2x-3) = 4+6 = 10
```

```python
#Inverse
a = np.array([[1,2],[3,4]])  # 1/|A| * adj(A) = 1 2 ----> adj ----> 4 -2
                                           #    3 4                -3  1
print("Inverse:\n",np.linalg.inv(a))
```

```python
# eigen--> a - lambda(i) #i - identity matrix, a --> given matrix
#i = 1 0 
#    0 1 
a = np.array([[1,2],[3,4]])
print(np.linalg.eig(a))
```

```python
#Rank Matrix
a = np.array([[2,24],[4,48]])
b = np.array([[1,2,3],[2,3,4],[3,5,7]])
print(a)
print(b)
print("Rank:",np.linalg.matrix_rank(a))
print("Rank:",np.linalg.matrix_rank(b))
```

```python
#Matrix Multiplication
x = np.array([1,2,3,4])
y = np.array([3,3,2,2])
print(np.dot(x,y)) #Same as matmul
print(np.matmul(x,y))
```

```python
#Solve
a = np.array([[4,3],[-5,9]])
b = np.array([[20],[26]])
np.linalg.solve(a,b)
```

```python
#Trace
a = np.array([[1,2],[3,4]]) #adds diagonally
print(np.trace(a))
b = np.array([[2,2],[3,6]])
print(np.trace(b))
```

### String Functions

```python
#String operation
#string information
#string comparisons
#rjust, ljust, lower, upper, split, join, strip, capitalize, title, center, encode, decode
import numpy as np
```

```python
#Lower and upper
a = 'Hello'
print(np.char.lower(a))
print(np.char.upper(a))
```

```python
#Split
a = "Hello World"
np.char.split(a)
```

```python
#Join
a = ["Hello","World"]
print(np.char.join("-",a))
b = "Hello World"
print(np.char.join(" ",b))
```

```python
#Capitalize
a = "hell"
np.char.capitalize(a)
```

```python
#Strip
a = "  Okay"
print(np.char.strip(a))
```

```python
#Title -- Same like capitalize
a = "hell"
print(np.char.title(a))
```

```python
#Center
a = "Help"
print(np.char.center(a,10,''))
```

```python
a = ['chef','maid']
z = np.char.encode(a, encoding = 'utf8', errors = None) #cp037, cp500
print(z)
```

```python
c = np.char.decode(z, encoding = 'utf8', errors = None)
print(c)
```

```python
#ljust
arr = np.array(["a","b","c"])
width = 10
np.char.ljust(arr,width,fillchar = '*')
```

```python
#rjust
arr = np.array(["a","b","c"])
width = 10
print(np.char.rjust(arr,width,fillchar = ''))
```
